```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:/Users/hp/Desktop/ml/areaprices.csv")
df
```

```python
%matplotlib inline
plt.scatter(df.area,df.prices)
```

```python
%matplotlib inline
plt.xlabel('area(sr ft)')
plt.ylabel('prices(US $)')
plt.scatter(df.area, df.prices, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.prices)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
