```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:/Users/hp/Desktop/ml/homeprices.csv")
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
%matplotlib inline
plt.xlabel('salary')
plt.ylabel('dependancy')
plt.scatter(df[['area','bedrooms','age']],df.price, color='blue', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```
